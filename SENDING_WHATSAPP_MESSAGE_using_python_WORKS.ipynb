```python
import pywhatkit as kit
import datetime

# List of WhatsApp numbers (include the country code in each number)
phone_numbers = [
    "+919748862503"
    # Add all 23 numbers here
]

message = "Hello, this is a test message sent from Python!"

# Get the current time
now = datetime.datetime.now()

# Schedule messages
for index, phone_number in enumerate(phone_numbers):
    try:
        # Schedule message for 2 minutes from now plus index as an offset to stagger messages
        send_time = now + datetime.timedelta(minutes=1 + index)
        kit.sendwhatmsg(phone_number, message, send_time.hour, send_time.minute)
        print(f"Message scheduled to {phone_number} at {send_time.hour}:{send_time.minute}")
    except Exception as e:
        print(f"Failed to schedule message to {phone_number}. Error: {str(e)}")
```
